```python
import pandas as pd
import matplotlib.pyplot as plt

# Set display options for pandas
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

# URL of the League of Legends champions list page
url = "https://leagueoflegends.fandom.com/wiki/List_of_champions"

# Use pandas to read the tables from the URL
tables = pd.read_html(url)

# Get the champions DataFrame from the list of tables
champions_df = tables[0]

# Drop duplicates based on the "Champion" column
champions_df.drop_duplicates(subset="Champion", keep="first", inplace=True)

# Split the "Champion" column into "Title" and "Name" columns
champions_df[["Title", "Name"]] = champions_df["Champion"].str.split(" ", 1, expand=True)

# Count the number of champions per class
class_counts = champions_df["Classes"].value_counts()

# Plot the data
plt.figure(figsize=(10, 6))
class_counts.plot(kind="bar")
plt.title("Number of Champions per Class")
plt.xlabel("Class")
plt.ylabel("Number of Champions")
plt.xticks(rotation=45, ha="right")

# Display the plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Set display options for pandas
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

# URL of the League of Legends champions list page
url = "https://leagueoflegends.fandom.com/wiki/List_of_champions"

# Use pandas to read the tables from the URL
tables = pd.read_html(url)

# Get the champions DataFrame from the list of tables
champions_df = tables[0]

# Drop duplicates based on the "Champion" column
champions_df.drop_duplicates(subset="Champion", keep="first", inplace=True)

# Split the "Champion" column into "Title" and "Name" columns
champions_df[["Title", "Name"]] = champions_df["Champion"].str.split(" ", 1, expand=True)

# Convert RP and Blue Essence columns to strings and then to numeric
champions_df["RP"] = pd.to_numeric(champions_df["RP"].astype(str).str.replace(",", ""), errors="coerce")
champions_df["Blue Essence"] = pd.to_numeric(champions_df["Blue Essence"].astype(str).str.replace(",", ""), errors="coerce")

# Filter out rows with missing RP or Blue Essence values
filtered_champions = champions_df.dropna(subset=["RP", "Blue Essence"])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_champions["Blue Essence"], filtered_champions["RP"], color="b", alpha=0.5)
plt.title("RP vs Blue Essence Cost of Champions")
plt.xlabel("Blue Essence")
plt.ylabel("RP")
plt.grid(True)

# Display the scatter plot
plt.tight_layout()
plt.show()



```
